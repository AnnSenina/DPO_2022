```python
import numpy as np
import random

with open('/Users/marinapozidaeva/Downloads/59495692-2.txt') as f:
    texts = f.read()

corpus = texts.split()

def make_pairs(corpus):
    for i in range(len(corpus)-1):
        yield (corpus[i], corpus[i+1])
        
ngrams = make_pairs(corpus)

word_dict = {}

for word_1, word_2 in ngrams:
    if word_1 in word_dict.keys():
        word_dict[word_1].append(word_2)
    else:
        word_dict[word_1] = [word_2]
```

```python
word_dict2 = {}

for k, v in word_dict.items():
    counter_dict = Counter(v)
    
    sorted_freqs = sorted(Counter(counter_dict).items(), key=lambda x:x[1], reverse=True)[:3]
    word_dict2[k] = [i[0] for i in sorted_freqs]
```

```python
def genearate_text():
    n_words = random.randint(10, 50)

    first_word = np.random.choice(corpus)

    while first_word.islower():
        first_word = np.random.choice(corpus)
    chain = [first_word]
    for i in range(n_words):
        chain.append(np.random.choice(word_dict[chain[-1]]))

    print(' '.join(chain))
```
